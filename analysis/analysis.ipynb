```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

actuals = pd.read_csv('data/actuals_jan.csv')
actuals['startTime'] = pd.to_datetime(actuals['startTime'])

# The prompt asks for "generation" column
wind_gen = actuals['generation']

p95_reliability = wind_gen.quantile(0.05)
mean_gen = wind_gen.mean()

print(f"Mean Generation: {mean_gen:.2f} MW")
print(f"Reliable Generation (P95): {p95_reliability:.2f} MW")

plt.figure(figsize=(10, 6))
sns.histplot(wind_gen, kde=True, color='blue')
plt.axvline(p95_reliability, color='red', linestyle='--', label=f'P95: {p95_reliability:.2f}MW')
plt.title('Reliability Analysis: Wind Generation Distribution')
plt.xlabel('MW')
plt.legend()
plt.show()
```
